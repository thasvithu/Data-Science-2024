```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load data set
df = pd.read_csv("result.csv")
```

```python
sns.countplot(x = df["Grade"])
```

```python
sns.histplot(x = df["Grade"])
```

```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the data
# Plotting the distribution of grades
plt.figure(figsize=(10, 6))
df['Grade'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Grades')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie chart of grade distribution
plt.figure(figsize=(8, 8))
df['Grade'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=140)
plt.title('Grade Distribution')
plt.ylabel('')
plt.show()
```

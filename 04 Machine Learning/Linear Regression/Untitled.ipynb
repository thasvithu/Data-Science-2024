```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("score.csv")
```

```python
df.head()
```

```python
x = df["Hours"].values.reshape(-1, 1)
y = df["Scores"].values
```

```python
np.shape(x)
```

```python
plt.scatter(x, y, color = "blue")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.title("Before Prediction")
```

```python
regModel = LinearRegression()
```

```python
regModel.fit(x, y)
```

```python
plt.scatter(x, y, color = "blue")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.title("After Prediction")
plt.plot(x, regModel.predict(x), color = 'red')
```

```python
x_new = np.array([[6.6]])
result = regModel.predict(x_new)
```

```python

plt.scatter(x, y, color = "blue")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.title("New_X Prediction")
plt.plot(x, regModel.predict(x), color = 'red')

plt.scatter(x_new, result, color = "green")
```
